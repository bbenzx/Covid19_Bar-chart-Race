# Read Confirm_case to Dataframe

# Read Death_case.csv to Dataframe

```python
death_case = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Covid19 Data Manipulation/time_series_covid19_deaths_global.csv')
death_case
```

# Read Recovery_case.csv to Dataframe

```python
recov_case = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Covid19 Data Manipulation/time_series_covid19_recovered_global.csv')
recov_case
```

# Append new column named "Confirmd_case"

```python
con_case['file'] = "Confirmed_case"
con_case
```

# Append new column named "Death_case"

```python
death_case['file'] = "Death_case"
death_case
```

# Append new column named "Recovered_case"

```python
recov_case['file'] = "Recoverd_case"
recov_case
```

# Concatenate all dataframe together

```python
all_data = pd.concat([con_case, death_case, recov_case], axis = 0)
all_data
```

# Read all columns in all_data dataframe

```python
all_data.columns
```

# Drop unwanted columns

```python
country_lv = all_data.drop(columns=['Province/State','Lat','Long'])
country_lv
```

# Group dataframe by "Country/Region" and "file" and sum all rows which have the same "country/region" and "file"

```python
country_lv = country_lv.groupby(by = ['Country/Region','file']).sum().reset_index()
country_lv
```

# Get all columns into list except "Country/Region" and "file"

```python
select_col_date = [col for col in country_lv.columns if col not in('Country/Region','file')]
select_col_date
```

# Unpivot

```python
country_lv_melt = country_lv.melt(id_vars=['Country/Region','file'], value_vars = select_col_date)
country_lv_melt
```

# Rename columns

inplace=True replace on the old dataframe, dont have to create a new one.

```python
country_lv_melt.rename(columns={'file':'Type_of_case', 'variable':'Date', 'value':'Number_of_case'},inplace=True)
country_lv_melt
```

# Get latest data date from List

```python
latest_date = select_col_date[-1]
latest_date
```

# Replace "/" in date with "."

```python
latest_date = select_col_date[-1].replace('/','.')
latest_date
```

# Export to Excel file

```python
country_lv_melt.to_csv(f'/content/drive/MyDrive/Colab Notebooks/Covid19 Data Manipulation/covid19_{latest_date}.csv',index=False)
```
